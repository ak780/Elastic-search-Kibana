```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch(HOST="http://localhost", PORT=9200)
es = Elasticsearch()
```

```python
import psutil
import time
flist = list()
i=0
for proc in psutil.process_iter():
   print(proc)
   pInfoDict = proc.as_dict(attrs=['pid', 'cpu_percent','memory_percent'])
   p = psutil.Process(pInfoDict['pid']).as_dict(attrs = ['create_time'])
   pc =  time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(p['create_time']))
   pInfoDict['time'] = pc
   print(pInfoDict)
   flist.append(pInfoDict)
   res = es.index(index="test-index-final", id=i, body=pInfoDict)
   i+=1
```

```python
for i in range(0,len(flist)):
    res = es.get(index="test-index-final", id=i)
    print(res)

```
